```python
import mne
import numpy as np
import os
import warnings
warnings.filterwarnings('ignore')

data_path = r"C:\Users\hiro2\Documents\EEG_project\data\MUSIN-G"
save_path = r"C:\Users\hiro2\Documents\EEG_project\data\timeseries"
os.makedirs(save_path, exist_ok=True)

MUSE_CH = [50, 26, 88, 7]
SF = 250
TARGET_LEN = 250 * 30  # 30秒分に統一

for sub_id in range(1, 21):
    for ses_id in range(1, 13):
        set_path = os.path.join(
            data_path,
            f"sub-{sub_id:03d}",
            f"ses-{ses_id:02d}",
            "eeg",
            f"sub-{sub_id:03d}_ses-{ses_id:02d}_task-MusicListening_run-{ses_id}_eeg.set"
        )
        if not os.path.exists(set_path):
            continue

        try:
            raw = mne.io.read_raw_eeglab(set_path, preload=True, verbose=False)
            raw.filter(1., 40., fir_window='hamming', verbose=False)
            data = raw.get_data()[MUSE_CH].astype(np.float32)  # (4, timepoints)

            # 長さを統一（短ければゼロパディング、長ければトリミング）
            if data.shape[1] >= TARGET_LEN:
                data = data[:, :TARGET_LEN]
            else:
                pad = np.zeros((4, TARGET_LEN - data.shape[1]), dtype=np.float32)
                data = np.concatenate([data, pad], axis=1)

            # 正規化
            data = (data - data.mean(axis=1, keepdims=True)) / \
                   (data.std(axis=1, keepdims=True) + 1e-8)

            fname = f"sub{sub_id:03d}_ses{ses_id:02d}.npy"
            np.save(os.path.join(save_path, fname), data)
            print(f"✓ sub-{sub_id:03d} ses-{ses_id:02d} shape={data.shape}")

        except Exception as e:
            print(f"✗ sub-{sub_id:03d} ses-{ses_id:02d}: {e}")

print(f"\n完了！shape=(4, {TARGET_LEN})")
```
